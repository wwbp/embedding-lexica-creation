```python
from copy import deepcopy

import numpy as np
import pandas as pd
```

We input tokens, scores, and two thresholds for highlighting.

```python
def highlight(tokens, scores, max_th, min_th):
    cur = deepcopy(tokens)
    pos_prefix = '''<span class="pos">'''
    neg_prefix = '''<span class="neg">'''
    suffix = '''</span>'''
    br = '<br>'
    scores = np.array(scores)
    pos_index = np.where(scores>max_th)[0]
    neg_index = np.where(scores<min_th)[0]
    for i in range(0,len(cur)):
        if i in pos_index:
            cur[i] = pos_prefix+cur[i]+suffix
        if i in neg_index: 
            cur[i] = neg_prefix+cur[i]+suffix
    html = ' '.join(np.append(cur,br))
    return html
```

```python
threshold = [[0.0906, -0.0817], [0.0898, -0.107], [0.168, -0.0654], [0.0215, -0.036]]
methods = ['Attention_Masking', 'Partition_SHAP', 'Integrated_Gradient', 'Lexica']
sentences=[]
for i in range(200):
    sentence = []
    subset = yelp[yelp['Unnamed: 0']==i]
    tokens = list(subset['Unnamed: 1'])
    for method, thre in zip(methods, threshold):
        scores = list(subset[method])
        #print(tokens)
        sentence.append(highlight(tokens, scores, thre[0], thre[1]))
    sentences.append(sentence)
```

```python
def calculate_mean_ck_method(df, yelp_str2id, yelp_index):
    ck = [[] for _ in range(4)]
    for hit in df['HITId'].unique():
        sub = df[df['HITId'] == hit]
        num = str(yelp_str2id[sub.iloc[0]['Input.sent0-method0']])
        word_col = [[] for _ in range(4)]
        for sent in range(11):
            for method in range(4):
                if sent == yelp_index[num][0]:
                    continue
                word_col[yelp_index[num][1][sent][method]].append('Answer.s'+str(sent)+'m'+str(method)+'-radios')
        for i in range(len(sub)-1):
            for j in range(i+1, len(sub)):
                for k in range(4):
                    score = cohen_kappa_score(sub.iloc[i][word_col[k]].astype(int), sub.iloc[j][word_col[k]].astype(int), labels=[1,2,3])
                    if np.isnan(score):
                        score = 1
                    ck[k].append(score)
    for i in range(4):
        print('Average CK:{:.3f}'.format(np.mean(ck[i])))
```
